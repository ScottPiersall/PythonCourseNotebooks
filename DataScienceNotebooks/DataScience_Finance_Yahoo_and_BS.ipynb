```python
#
# Data Science with Yahoo! Finanace and Some HTML scraping using beautiful sou
#
#
import yfinance as yf
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt


start_date = dt.datetime(2024,1,1)
end_date = dt.datetime(2025,3,31)




```

```python
data = yf.download('TSLA', start=start_date, end=end_date)
```

```python
data.head(15)
```

```python
data.reset_index( inplace=True)

after_date = dt.datetime(2025,1,1)

#filtereddf = data['Date'] > after_date

filtereddf = data.loc[ data['Date'] > after_date    ]




```

```python
filtereddf.head(10)
```

```python




plt.figure( figsize = (40,20))
plt.title("TESLA Closing Prices from {} to {}".format(start_date, end_date))
plt.plot(filtereddf['Close'])
plt.plot(filtereddf['Open'])
plt.plot(filtereddf['High'])
plt.plot(filtereddf['Low'])
plt.show
```
