```python
class Shape:
    def __init__(self, name="", sides=3):
        self.name = name
        self.sides = sides

    def __str__(self):
        return "Shape name: " + self.name + ", Sides : " + str(self.sides)
    

class Square(Shape):
    def __init__(self, name="", sidelength=1):
        super().__init__(name, 4)
        self.sidelength = sidelength

    def Area(self):
        return self.sidelength * self.sidelength

class Rectangle(Shape):
    """
    This is a specific implementation of a Rectangle using shape

    """
    def __init__(self, name="", width=1, height=1):
        super().__init__(name, 4)
        self.width = width
        self.height = height

    def Area(self):
        """
        Returns the area of the rectangle

        
        Parameters:
            none

        Returns:
        int:  The area (length x width) of this rectangle    
        """
        return self.width * self.height



```
